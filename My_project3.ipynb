```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('https://www.numbeo.com/health-care/rankings_by_country.jsp','html.parser')
soup = requests.get(url)
soup.text

soup = BeautifulSoup(soup.text,"lxml")
soup
```

```python
table_code = soup.table
table_code
```

```python
tags = table_code.find_all('td')
tags
```

```python
data = []
for a in tags:
    data.append(a.text)
    
data
```

```python
data = []
for a in tags:
    data.append(a.text.split('\n'))
    
data
```

```python
import csv
file = open('health_data.csv','w')
x = csv.writer(file)
x.writerows(data)
file.close()
```
