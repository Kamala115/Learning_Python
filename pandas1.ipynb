```python
import pandas as pd
df1 = pd.DataFrame({'Name':['Ram','Shyam','Hari','Rameshwor','Pandit'],
                   'Age':[12,21,34,70,60]})
df2 = pd.DataFrame({'Name':['Rabi','Sabi','Habi','Saraswati','Chandra'],
                   'Age':[21,41,54,35,22]})
df3 = pd.DataFrame({'Name':['Chunu','Shunu','Hunu','Kabita','Kamala'],
                   'Age':[50,20,10,40,23]})

df = pd.concat([df1,df2,df3],keys = ['df1','df2','df3'],names = ['dataframes','Row_num'])
df.to_csv('census.csv',index = False)
df
```

```python
import pandas as pd
df1 = pd.DataFrame({'Name':['Ram','Shyam','Hari','Rameshwor','Pandit'],
                   'Age':[12,21,34,70,60]})
df2 = pd.DataFrame({'Name':['Rabi','Sabi','Habi','Saraswati','Chandra'],
                   'Age':[21,41,54,35,22]})
df3 = pd.DataFrame({'Name':['Chunu','Shunu','Hunu','Kabita','Kamala'],
                   'Age':[50,20,10,40,23]})

df = pd.concat([df1,df2,df3],keys = ['Group1','Group2','Group3'],names = ['Group','Row_num']).reset_index()
df.to_csv('census.csv',index = False)
df
```
